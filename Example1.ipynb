```python
from Kalman_Filter import Kalman_Filter
from generate_lds import *
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Dynamics
A = 0.9*np.random.rand(2,2)
B = np.zeros((2,1))
Q = 0.01*np.eye(2,2)

# Measurements
C = np.random.rand(1,2)
D = np.random.rand(1,1)
R = [0.01]

# Control input
U = np.zeros((1,100))

initX = np.zeros((2,1))
initV = 0.0001*np.eye(2,2)

X, Y = generate_lds(U, A, B, C, D, Q, R, initX, initV)

Xp, Pp, Xf, Pf, Kf = Kalman_Filter(Y, U, A, B, C, D, Q, R, initX, initV)
```
